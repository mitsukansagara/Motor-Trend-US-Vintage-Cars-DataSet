```python
#define the path, please note you need to use your own path info for mtcars.csv to read the data
automobile = pd.read_csv(r'D:\Mitsu\MS-ISM\508 - Data Mining I\Week2\mtcars-1.csv')
```

```python
# select numerical columns that are used to perform cluster analysis
X= automobile[['mpg','cyl','disp','hp','drat','wt','qsec','gear','carb']]
```

```python
# standardize features
from sklearn.preprocessing import StandardScaler

#create a scaler
scaler = StandardScaler()

# Apply the scaler
automobile_std = scaler.fit_transform(X)
automobile_std
```

```python
#Create dendrogram using average linkage
import scipy.cluster.hierarchy as shc
import matplotlib.pyplot as plt
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
#average linkage is used here - minimum average distance between two centorids
linked = linkage(automobile_std, 'average')
plt.figure(figsize=(15, 10))
dendrogram(linked, orientation='top', labels = np.array(automobile['model']), distance_sort='descending')
plt.ylabel('Euclidean Distance')
plt.xlabel('Model')
plt.title('Dendrogram using Average Linkage')
plt.show()
```

```python
#Create dendrogram using ward linkage
#safest method to use is ward while safest algo is k-means
import scipy.cluster.hierarchy as shc
import matplotlib.pyplot as plt
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
#ward linkage is used here 
linked = linkage(automobile_std, 'ward')
plt.figure(figsize=(15, 10))
dendrogram(linked, orientation='top', labels = np.array(automobile['model']), distance_sort='descending')
plt.ylabel('Euclidean Distance')
plt.xlabel('Model')
plt.title('Dendrogram using Ward Linkage')
plt.show()
```
